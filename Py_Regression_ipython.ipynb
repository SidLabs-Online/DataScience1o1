## Linear Regression 
Linear Regression is the most important and widely known Machine Learning Algorithms which people start with. A linear regression model consists of:

Discreet/continuous independent variables
A best-fit regression line
Continuous dependent variable. The equation of the Linear Regression is: 
# `ŷ = b0 + b1x`
__where__:<br>
- ŷ: The estimated response value
- b0: The intercept of the regression line
- b1: The slope of the regression line

The equation above is used to predict the value of the target based on the given predictors.

### Problem Statement
The attached sample data (`Sale_SL_2021.csv`) consists of rows and columns of expenses made by a firm SidLabs on its marketing and promotions. The objective of this analysis is to predict how the expenses made are/are not optimizing the sales.

Also, with this analysis it is expected to derive the next problem to solve in order to optimize sales for SidLabs.



```python
# making the necessary imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data= pd.read_csv('Sale_SL_2021.csv') # Reading the data file
data.head() # checking the first five rows from the dataset
```

#### What do you see ? what are the features?
LinkedIn : Dollars spent for advertising on TV for a single product in a given market (in thousands of dollars)
Radio: Dollars spent for advertising on Radio
SEO : Dollars spent for advertising on Google Ads
Newspaper: Dollars spent for advertising on Newspaper

#### What is the response?
Sale: Sale of a single product in a given market (in thousands)

```python
# using the pandas dataframe.shape method 
print("The size of this data set :\n")
print('\tRows : ', data.shape[0], 
      "\n\tColumns : ",data.shape[1] ,"\n")

print("Printing the summary of this data set :\n>>")
data.info()
```

```python
# visualizing the relationship between features and response using scatterplots
fig, axs = plt.subplots(1, 4, sharey=True)
data.plot(kind='scatter', x='LinkedIn', y='Sale', ax=axs[0], figsize=(16, 8))
data.plot(kind='scatter', x='Radio', y='Sale', ax=axs[1])
data.plot(kind='scatter', x='SEO', y='Sale', ax=axs[2])
data.plot(kind='scatter', x='News', y='Sale', ax=axs[3])
```

## Perform Simple Linear Regression
Once we’ve confirmed that the relationship between our variables is linear and that there are no outliers present, we can tryy testing to fit a simple linear regression model using '_News_' as the explanatory variable and '_Sale_' as the response variable:

Note: We’ll use the OLS() function from the statsmodels library to fit the regression model.

```python
import statsmodels.api as sm

#define response variable
y = data['Sale']

#define explanatory variable
x = data[['News']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```

Here is how to interpret the summary:

- P>|t|: This is the p-value associated with the model coefficients. Since the p-value for News (0.125) is significantly higher than .05, we can say that there is __NO__ statistically significant association between News and Sale.
- R-squared: This number tells us the percentage of the variation in the exam scores can be explained by the number of hours studied. In general, the larger the R-squared value of a regression model the better the explanatory variables are able to predict the value of the response variable. In this case, only 4.8% of the variation in Sale can be explained by spends on News.
- F-statistic & p-value: The F-statistic (2.434) and the corresponding p-value (0.125) tell us the overall significance of the regression model, i.e. whether explanatory variables in the model are useful for explaining the variation in the response variable. Since the p-value in this example is greater than .05, our model is statistically in-significant and News is deemed to be not useful for explaining the variation in Sale.

```python
#define figure size
fig = plt.figure(figsize=(12,8))

#produce residual plots
fig = sm.graphics.plot_regress_exog(model, 'News', fig=fig)
```

```python
#define residuals
res = model.resid

#create Q-Q plot
fig = sm.qqplot(res, fit=True, line="45")
plt.show()
```
